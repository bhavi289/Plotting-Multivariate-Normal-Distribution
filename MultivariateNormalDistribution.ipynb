```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import multivariate_normal
```

```python
def g(mu, cov, prior, pos):
    rv = multivariate_normal(mu, cov)
    Z = rv.pdf(pos) * prior
    fig = plt.figure(figsize=(10, 15))
    ax = fig.add_subplot(211, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.contour(X, Y, Z,offset=0.03, zdir='z')
    ax.set_title("Mean = " + str(mu) + ", Covariance Matrix = " + str(cov) + ", Prior = " + str(prior)+"\n")

    
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('p')
    
#     ax = fig.add_subplot(212, projection='3d')
#     ax.contour(X, Y, Z,offset=0, zdir='z')
    
#     ax.set_xlabel('x1')
#     ax.set_ylabel('x2')
#     ax.set_zlabel('p')
```

```python
x = np.linspace(-5, 9, 50)
y = np.linspace(-5, 15, 50)

X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

d = 2

mu_1 = np.array([0, 0])
mu_2 = np.array([4, 10])

cov_1 = np.array([[1, 0], [0,1]])
cov_2 = np.array([[3, 1], [1,3]])

prior_1 = 0.5
prior_2 = 0.5

g(mu_1, cov_1, prior_1, pos)
g(mu_2, cov_2, prior_2, pos)

rv = multivariate_normal(mu_1, cov_1)
Z1 = rv.pdf(pos) * prior_1

rv = multivariate_normal(mu_2, cov_2)
Z2 = rv.pdf(pos) * prior_2

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z1)
# ax.contourf(X, Y, Z1,offset=0.035, zdir='z')

ax.plot_wireframe(X, Y, Z2)
# ax.contourf(X, Y, Z2,offset=0.04, zdir='z')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('p')
```
